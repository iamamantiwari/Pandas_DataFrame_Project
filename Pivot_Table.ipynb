```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('dataset/weather.csv')
```

```python
df.pivot(index='city',columns='date')
```

```python
df.pivot(index='city',columns='date',values='temperature')
```

```python
df.pivot(index='date',columns='city',values='humidity')
```

```python
df2=pd.read_csv('dataset/weather2.csv')
```

```python
df2
```

```python
df2.pivot_table(index='city',columns='date')
```

```python
df2.pivot_table(index='city',columns='date',aggfunc=np.sum)
```

```python
df2.pivot_table(index='city',columns='date',aggfunc=np.max,margins=True,margins_name='Sum')
```

```python
df2.pivot_table(index='city',columns='date',aggfunc=np.max,margins=True,margins_name='Sum',values='humidity')
```

```python
df2.pivot_table(index='city',columns='date',aggfunc=np.max,margins=True,margins_name='Sum',values='temperature')
```
